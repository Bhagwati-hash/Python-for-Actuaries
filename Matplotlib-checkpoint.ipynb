# Customization of plots

```python
titanic.Fare.plot(figsize=(12,8),c='blue',fontsize=14,linestyle='-')
```

```python
titanic.Age.plot(figsize=(12,6),c='r')
plt.title('Titanic-Ages',fontsize=15)
plt.xlabel('passanger id',fontsize=14)
plt.ylabel('ages',fontsize=14)
plt.legend(fontsize=12)
```

```python
#how to change the limits for x axis and y axis
titanic.Age.plot(figsize=(12,6),c='r')
plt.title('Titanic-Ages',fontsize=15)
plt.xlabel('passanger id',fontsize=14)
plt.xlim(200,400)
plt.ylim(0,50)
```

```python
#if we want to change the gaps between x axis and y axis
xnewticks=[x for x in range(0,401,50)]
ynewticks=[x for x in range(0,80,5)]
titanic.Age.plot(figsize=(12,6),xlim=(0,400),ylim=(0,80),xticks=xnewticks,yticks=ynewticks,c='g',xlabel='passanger id',ylabel='age',rot=45)
```
